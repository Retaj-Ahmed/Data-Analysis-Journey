```python
import pandas as pd

# 1. إنشاء البيانات (كأننا بنعمل جدول درجات)
students_data = {
    'Name': ['Retaj', 'Remass', 'Adam', 'Zain', 'Ahmed'],
    'Age': [18, 16, 11, 7, 41],
    'Score': [90, 50, 25, 30, 78]
}

# 2. تحويل البيانات لـ DataFrame (الجدول اللي بنحلله)
df = pd.DataFrame(students_data)

# 3. تصفية البيانات (عايزين بس اللي نجحوا وجابوا أكتر من 50)
successful_students = df[df['Score'] > 50]

print("--- All Students ---")
print(df)

print("--- Successful Students Only ---")
print(successful_students)
```

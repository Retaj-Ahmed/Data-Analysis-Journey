```python
import pandas as pd
```

```python
people = {
    "first": ["Retaj", "Remass", "Marwa"],   
    "second": ["Ahmed", "Ahmed", "Salah"],  
    "email": ["retagahmed@gmail.com", "ahmedadel@gmail.com", "marwasalah@gmail.com"]
}
```

```python
df = pd.DataFrame(people)
df
```

```python
#comparison
filt = df['second'] == 'Ahmed'
filt
```

```python
df[filt]
```

```python
#OR
df[df['second'] == 'Ahmed']
```

```python
#OR
df.loc[filt]
```

```python
df.loc[filt,'email']
```

```python
filt = (df['second']== 'Ahmed') & (df['first'] == 'Retaj')
filt
```

```python
df.loc[filt,'email']
```

```python
filt = (df['second']== 'Ahmed') | (df['first'] == 'Retaj')
filt
```

```python
df.loc[filt,'email']
```

```python
#Negate that filter (opposite results)
df.loc[~filt,'email']
```
